# Demonstracija i plottanje Rezultata

## 40 x 60 pixels regije bez background klase

```python
import csv
import numpy as np
from sklearn.metrics import accuracy_score, f1_score


#na koliko PRVIH slika u test skupu radimo testiranje (--> na lastImg_index +1) ?
lastImg_index = 9999


#stvaramo vektor predikcija
predictions  = []

with open('DobarModel25klasa/Najbolji90Average10k.csv', 'rt') as csvfile:
    spamreader = csv.reader(csvfile)
    i = 0
    for row in spamreader:     
        if(i>lastImg_index):
            break
        predictions.append(row)
        i = i+1       

#stvaramo vektor ground truth labela
groundTruth_labels = []
with open('groundTruthInitialTestSet.csv', 'rt') as correct:
    spamreader = csv.reader(correct)
    i = 0
    for row in spamreader:
        if(i>lastImg_index):
            break
        groundTruth_labels.append(row)
        i = i+1
unique_elements, counts_elements = np.unique(groundTruth_labels, return_counts=True)

acc = accuracy_score(groundTruth_labels, predictions, normalize=True, 
                               sample_weight=None)
f1macro = f1_score(groundTruth_labels, predictions,labels=unique_elements, average='macro')
f1micro = f1_score(groundTruth_labels, predictions,labels=unique_elements, average='micro')
f1weighted = f1_score(groundTruth_labels, predictions,labels=unique_elements, average='weighted')
f1each=f1_score(groundTruth_labels, predictions,labels=unique_elements, average=None)
#f1samples = f1_score(groundTruth_labels, predictions,labels=unique_elements, average='samples')
print("Accuracy = " + str(acc) )
print("F1Macro = " + str(f1macro))
print("F1Micro = " + str(f1micro))
print("F1Weighted = " + str(f1weighted))
print("\n")
print("F1 For Each class:")
print("\n")
for i in range(0,len(unique_elements)):
    print("F1Score " + unique_elements[i] + " = " + str(f1each[i]))
#print(unique_elements)

f =  open('acc_f1Najbolji90Average10k.txt', 'w+') 
f.write("Accuracy: %f\n" %acc)
f.write("F1 score macro: %f\n" %f1macro)
f.write("F1 score micro: %f\n" %f1micro)
f.write("F1 score weighted: %f\n" %f1weighted)
#f.write("F1 score: %f\n" %f1samples)
f.close()
```

```python
import matplotlib.pyplot as plt; plt.rcdefaults()
import matplotlib.pyplot as plt

y_pos = np.arange(len(unique_elements))
#print(y_pos)

plt.bar(y_pos, f1each, align='center', alpha=0.5)
plt.xticks(y_pos, unique_elements,rotation='vertical')
#plt.yticks(xrange(10), rotation='vertical')
plt.ylabel('F1 Score')
plt.title('No-Background-Class')

plt.savefig('Nobackground10k.png')
plt.show()
#plt.savefig('Nobackground10k.png')        
plt.close()
```
